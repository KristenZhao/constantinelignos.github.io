```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Use read_csv to load a CSV file as a DataFrame
ratings = pd.read_csv("http://lignos.org/pybootcamp/examples/presidents.csv")
```

```python
ratings['year_round'] = ratings.year.astype(int)
```

```python
ratings.head()
```

```python
ratings_year_means = ratings.groupby('year_round', as_index=False).mean()
```

```python
sns.pointplot(x="year_round", y="rating", data=ratings_year_means)
```
